```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Weather dataset.csv')

print(df.head())

print("\nDataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)
```

```python
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())
```

```python
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:")
print(df.duplicated().sum())
```

```python
print(df.columns[0])
```

```python
df[df.columns[0]] = pd.to_datetime(
    df[df.columns[0]],
    errors='coerce'
)

print(df.head())
print(df.info())
```

```python
date_col = df.columns[0]

plt.figure(figsize=(10,5))
plt.plot(df[date_col], df['Temp_C'])

plt.title("Temperature Trend Over Time")
plt.xlabel("Date/Time")
plt.ylabel("Temperature (°C)")
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['Rel Hum_%'], bins=20)

plt.title("Humidity Distribution")
plt.xlabel("Relative Humidity (%)")
plt.ylabel("Frequency")

plt.show()
```

```python
weather_counts = df['Weather'].value_counts().head(10)

weather_counts.plot(
    kind='bar',
    figsize=(10,5)
)

plt.title("Top 10 Weather Conditions")
plt.xlabel("Weather Type")
plt.ylabel("Count")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df['Wind Speed_km/h'], bins=20)

plt.title("Wind Speed Distribution")
plt.xlabel("Wind Speed (km/h)")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    df['Temp_C'],
    df['Rel Hum_%']
)

plt.title("Temperature vs Humidity")
plt.xlabel("Temperature (°C)")
plt.ylabel("Humidity (%)")

plt.show()
```
